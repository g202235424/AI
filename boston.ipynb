```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston
```

```python
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df
```

```python
df['PRICE'] = boston.target
```

```python
df
```

```python
# 2x4 subplot 이용. axs는 4x2
fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)
9
lm_features = ['LSTAT', 'RM', 'PTRATIO', 'INDUS']

# i에는 인덱스가 feature에는 RM ~ RAD까지 순차적으로 들어감
for i, feature in enumerate(lm_features):
    row = int(i/4) # 2행
    col = i%4

    # sns.regplot : 회귀직선을 그려줌
    sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

x_data=df.loc[:, ['LSTAT', 'RM', 'PTRATIO', 'INDUS']]
y_data=df.loc[:, 'PRICE']
X_train, X_test, y_train, y_test=train_test_split(x_data,
                                                  y_data,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# 메소드체이닝(chaining)을 사용하여 fit과 transform을 연달아 호출합니다
X_scaled = scaler.fit(X_train).transform(X_train)

# 위와 동일하지만 더 효율적입니다(fit_transform)
X_scaled_d = scaler.fit_transform(X_train)

#해당 fit으로 test데이터도 transform 해줍니다
X_test_scaled = scaler.transform(X_test)

plt.scatter(X_scaled[:,0],X_scaled[:,1])
```

```python
sns.pairplot(df,
             vars=['LSTAT', 'RM', 'PTRATIO', 'INDUS'],
            hue ='PRICE')
plt.show()
```

```python
#상관 관계 행렬
df_corr=df.corr()

#히트맵
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

accuracy1 = lr.score(X_train, y_train)
accuracy = lr.score(X_test, y_test)

print(accuracy)
print(accuracy1)
```

```python
print(lr.score(train_poly, y_train))
print(lr.score(test_poly, y_test))
```
